```python
lat_lon = pd.read_csv('locations_lat_lon.csv')
```

```python
lat_lon['location'] = lat_lon['name']
```

```python
lat_lon = lat_lon.drop(['name'], axis = 1)
```

```python
lat_lon
```

```python
df
```

```python
result = pd.merge(df, lat_lon, on=["location"])
```

```python
result
```

```python
heat_map_cols = result[['lat', 'lon', 'label']]
heat_map_cols.drop_duplicates(['lat', 'lon'], inplace=True)
```

```python
pivot = heat_map_cols.pivot(index='lat', columns='lon', values='label')
```

```python
sns.heatmap(pivot,  annot=True)
```
